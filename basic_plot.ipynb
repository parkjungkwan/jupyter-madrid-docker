```python
import matplotlib.pyplot as plt
```

```python
def plot_show():
    plt.title("plotting")
    plt.plot([10, 20, 30, 40])
    plt.show()
```

```python
plot_show()
```

```python
def plot_two_list_show():
    plt.plot([1, 2, 3, 4], [12, 43, 25, 15])
    plt.show()
```

```python
plot_two_list_show()
```

```python
def plot_marker():
    plt.plot(range(100))
    plt.show()
```
